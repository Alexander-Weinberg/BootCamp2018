# Question 1

```python
import numpy as np
import matplotlib.pyplot as plt

@np.vectorize
def problem1(n=int):
    
    mat = np.random.normal(size=(n,n))

    avgs = np.mean(mat, axis=1)
    v = np.var(avgs)
    return v
```

```python
def part2(step=100):
    grid = np.arange(100,1100,step)
    results = problem1(grid)
    plt.plot(grid,results)
```

```python
part2()
```

# Question 3

```python
def f(x):
    return 1 / (x - 1)

def plot_f():
    domainL = np.linspace(-2, 1-.15, 50)
    domainR = np.linspace(1+.15, 6, 50)
    plt.plot(domainL,f(domainL),color='magenta',ls='--',lw=4)
    plt.plot(domainR,f(domainR),color='magenta',ls='--',lw=4)
    plt.xlim(-2,6)
    plt.ylim(-6,6)
    plt.show()

plot_f()
```

# Question 4

```python
domain = np.linspace(0,2*np.pi,100)
codomain = [np.sin(domain), np.sin(2*domain), 2*np.sin(domain), 2*np.sin(2*domain)]
titles = [r'sin(x)', r'sin(2x)',r'2sin(x)',r'2sin(2x)']
styles = ['g-', 'r--', 'b--', 'm:']


fig, axes = plt.subplots(2, 2)
axes = np.ravel(axes,'C')
for i, ax in enumerate(axes):
    ax.plot(domain,codomain[i],styles[i])
    ax.set_title(titles[i])


plt.axis([0, 2*np.pi, -2, 2])
fig.suptitle('Lotta functions')
plt.tight_layout()
```

# Question 5

```python
fars = np.load('FARS.npy')

time = fars[:,0]
lat = fars[:,1]
long = fars[:,2]
```

```python
fig, axes = plt.subplots(1,2)

axes[0].plot(lat,long,'k,')
axes[0].set_xlabel('Latitude')
axes[0].set_ylabel('Longitude')
axes[0].set_aspect('equal')


axes[1].hist(time,bins=24, range=[0,24])
axes[1].set_xlabel('Hours')

plt.tight_layout()
```

# Question 6

```python
def g(x,y):
    ans = np.sin(x) * np.sin(y) / (x * y)
    return ans

domain = np.linspace(-2*np.pi, 2*np.pi, 100)
X, Y = np.meshgrid(domain,domain)
Z = g(X,Y)

#fig, axes = plt.subplots(2,1)

plt.subplot(211)
plt.pcolormesh(X,Y,Z,cmap='viridis')
plt.axis([-2*np.pi, 2*np.pi, -2*np.pi, 2*np.pi])
plt.colorbar()


plt.subplot(212)
plt.contour(X,Y,Z,10,cmap='coolwarm')
plt.axis([-2*np.pi, 2*np.pi, -2*np.pi, 2*np.pi])
plt.colorbar()


plt.tight_layout()
```
